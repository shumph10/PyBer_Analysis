```python
#import matplotlib dependency
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis, y_axis)
plt.plot()
```

```python
#create chart with object-oriented method
#create figure with:
fig, ax = plt.subplots()
#define the axes and put the axes inside the function
ax.plot(x_axis, y_axis, color="green",marker= "*",linewidth = 2, label= "Boston")
ax.set_ylim(0,45)
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.legend()
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create a bar graph with obj-oriented method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
#create a horizontal bar chart w/ obj o method
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
plt.gca().invert_yaxis()
ax.set_title('PyBer Fare by Month')
ax.set_ylabel('Date')
ax.set_xlabel('Fare($)')
ax.legend()
```

```python
#Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create a scatter plot based on x & y values above
#add 'o' so that you dont get a line graph or use scatter
# plt.plot(x_axis,y_axis, 'o')
plt.scatter(y_axis,x_axis, c="red", label="Chicago")
plt.xlabel('Total Fare($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.legend()
plt.gca().invert_yaxis()
plt.xlim(0,50)
```

```python
#scatter plot w/ ooi
fig,ax = plt.subplots()
ax.scatter(x_axis,y_axis, s = [i * 5 for i in y_axis], c="blue")
```

```python
#creating a pie chart with the MATLAB method
    #use one data array for the values of the slices of pie and the other for the labels for each slice
plt.subplots(figsize=(8, 8))
        ###HAS to be before the pie chart is made or itll make a sep graph
    #increase the size of the pie chart so %#s arent as crowded
plt.pie(y_axis, labels=x_axis)
```

```python
#can add %s, specific colors, or explode or "pop out" one or more wedges
    #to explode one put in 0 for reg slices in array then % you want popped out for the index of the slice you wanna pop
    #length of array has to = orig array length
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
    #autopct to add % and .1f% for 1 dec place, %before and after that formats # as a %
    #explode = array name created before
#increase the size of the pie chart so %#s arent as crowded in prev cell before pie chart is made
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
##makes an array for the colors and sets colors to it to change all colors, edits explode and % as before
```

```python
#pie charts with ooi
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize= (8,8))
ax.pie(y_axis, labels=x_axis, explode= explode_values, colors=colors, autopct='%.1f%%', shadow = True, startangle = 95,counterclock = False)
plt.show()
#formatting https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.pie.html
```
